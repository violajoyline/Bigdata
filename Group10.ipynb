```python
df = pd.read_csv("Tweets.csv")
df.head()
```

```python
import re
# Drop duplicates
df = df.drop_duplicates()

# Drop unused columns
df = df.drop(columns=['tweet_id', 'tweet_coord', 'tweet_location', 'user_timezone'], errors='ignore')

# Handle missing values
df['negativereason'].fillna('Unknown', inplace=True)

# Clean the text
def clean_text(text):
    text = re.sub(r"http\S+", "", str(text))     # remove URLs
    text = re.sub(r"@\w+", "", text)             # remove @mentions
    text = re.sub(r"#", "", text)                # remove hashtags symbol
    text = re.sub(r"\s+", " ", text)             # normalize spaces
    return text.strip()

df['clean_text'] = df['text'].apply(clean_text)

#  outlier check
df['tweet_length'] = df['text'].apply(len)

# Remove unusually long tweets
df = df[df['tweet_length'] < 300]

df.head()
```

```python
# Sentiment distribution
print("Sentiment Distribution:")
print(df['airline_sentiment'].value_counts())

# Tweets per airline
print("\nTweets Per Airline:")
print(df['airline'].value_counts())

# Sentiment per airline
print("\nSentiment per Airline:")
print(df.groupby(['airline', 'airline_sentiment']).size().unstack())

# Top negative reasons
print("\nTop Reasons for Negative Sentiment:")
print(df['negativereason'].value_counts())
```

```python
#overall sentiment
df['airline_sentiment'].value_counts().reset_index().rename(columns={
    'index': 'Sentiment',
    'airline_sentiment': 'Count'
}).to_csv('sentiment_distribution.csv', index=False)
```

```python
#negative tweets per airline
df[df['airline_sentiment'] == 'negative']['airline'].value_counts().reset_index().rename(columns={
    'index': 'Airline',
    'airline': 'Negative_Tweet_Count'
}).to_csv('neg_tweets_by_airline.csv', index=False)
```

```python
#most common reason for negative tweets
df['negativereason'].value_counts().dropna().reset_index().rename(columns={
    'index': 'Negative_Reason',
    'negativereason': 'Count'
}).to_csv('negative_reasons.csv', index=False)
```

```python
#customer service complaints
df[df['negativereason'] == 'Customer Service Issue']['airline'].value_counts().reset_index().rename(columns={
    'index': 'Airline',
    'airline': 'Customer_Service_Complaints'
}).to_csv('cust_service_issues.csv', index=False)
```

```python
#positive tweets per airline
df[df['airline_sentiment'] == 'positive']['airline'].value_counts().reset_index().rename(columns={
    'index': 'Airline',
    'airline': 'Positive_Tweet_Count'
}).to_csv('positive_tweets_by_airline.csv', index=False)
```

```python
#general overview
import pandas as pd

general_info = {
    'Total_Tweets': [df.shape[0]],
    'Unique_Airlines': [df['airline'].nunique()],
    'Unique_Users': [df['name'].nunique()],
    'Positive_Tweets': [df[df['airline_sentiment'] == 'positive'].shape[0]],
    'Neutral_Tweets': [df[df['airline_sentiment'] == 'neutral'].shape[0]],
    'Negative_Tweets': [df[df['airline_sentiment'] == 'negative'].shape[0]]
}

pd.DataFrame(general_info).to_csv('general_overview.csv', index=False)
```
